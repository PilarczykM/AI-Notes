# Baza wektorowa

## Czym jest baza wektorowa?
**Baza wektorowa** to specjalistyczny typ bazy danych, który przechowuje dane w formie wektorów, czyli matematycznych reprezentacji obiektów w przestrzeni wielowymiarowej. Wielowymiarowośc moze byc określana w setkach czy tysiącach wymiarów.

Każdy wektor może reprezentować różne cechy danych, takie jak `słowa`, `obrazy` czy `dźwięki`, co czyni je niezwykle przydatnymi w aplikacjach takich jak przetwarzanie języka naturalnego (NLP), systemy rekomendacyjne oraz semantyczne wyszukiwanie.

### Wektor

Wektory są zwane `embeddings`, odzwierciedla semantyczne znaczenie danej, która została "osadzona" z ang. embedded

Tworzenie wektorów z danych, `embedding`, o którym poniżej, to proces, który daje możliwość przeniesienia semantycznej zależności do świata liczb.

* Wyobraź sobie dwuwymiarową przestrzeń w formie wykresu z osią X i Y.
* Na przestrzeń nałożone są twoje dane. Ustalmy, że jest to 10 data pointów, jak z wykresu poniżej.
* Każda dana posiada odpowiednią liczbę na osi X i Y, czyli dwuwymiarowy wektor.

## Embedding model

Dowolny tekst – tweet, dokument lub książka – w jednej, zwartej reprezentacji.

Na tym polega siła `embedded modelu`, która leży u podstaw wielu systemów wyszukiwania.

Modele te przekształcają ludzki język w format, który maszyny mogą zrozumieć i porównać z szybkością i dokładnością.

Przyjmują tekst jako dane wejściowe i tworzą **tablicę liczb o stałej długości**, numeryczny odcisk palca semantycznego znaczenia tekstu.

`Embeddings` umożliwiają systemowi wyszukiwania znalezienie odpowiednich dokumentów nie tylko na podstawie dopasowań słów kluczowych, ale także zrozumienia semantyki.

![embedding concept](./images/embeddings_concept.png)

1) Embed text as a vector - `Embeddings` transformują tekst do reprezentacji numerycznego vektora.
2) Mesure similarity - `Embedding` wektory mogą być porównywane uzywając do tego operacji matematycznych.


```python
embeddings = embeddings_model.embed_documents(
    [
        "Hi there!",
        "Oh, hello!",
        "What's your name?",
        "My friends call me World",
        "Hello World!"
    ]
)
doc_count, embedding_count = len(embeddings), len(embeddings[0])
print(f"Document count: {doc_count}, Embedding count: {embedding_count}")
# Document count: 5, Embedding count: 1536
```

## Kluczowe cechy baz wektorowych
* **Zoptymalizowane przechowywanie**: Bazy wektorowe są zaprojektowane z myślą o **efektywnym przechowywaniu** i **wyszukiwaniu** wektorów o wysokiej wymiarowości. Używają wyspecjalizowanych struktur danych i algorytmów do zarządzania dużymi zbiorami danych.
* **Wydajne wyszukiwanie podobieństw**: Te bazy danych umożliwiają szybkie lokalizowanie wektorów na podstawie ich bliskości do podanego wektora zapytania, stosując miary podobieństwa, takie jak odległość euklidesowa czy podobieństwo cosinusowe.
* **Skalowalność**: Bazy wektorowe są architektonicznie przystosowane do poziomego skalowania, co pozwala na obsługę rosnącej ilości danych i zapytań bez znaczącego spadku wydajności.
* **Wsparcie dla osadzania**: Wektorowe bazy danych często przechowują osadzenia generowane przez modele uczenia maszynowego, co jest kluczowe dla reprezentacji danych w gęstej przestrzeni.
* **Przetwarzanie w czasie rzeczywistym**: Wiele z tych baz jest zoptymalizowanych do przetwarzania w czasie rzeczywistym, co czyni je idealnymi dla aplikacji wymagających szybkich odpowiedzi.

## Jak działają bazy wektorowe?
Bazy wektorowe działają na zasadzie wyszukiwania najbliższych sąsiadów **(Approximate Nearest Neighbor - ANN)**, co oznacza, że zamiast szukać dokładnych dopasowań, identyfikują najbliższe podobieństwa między wektorami. Proces konwersji danych na wektory, znany jako "embedowanie", może być realizowany za pomocą różnych technik, w tym modeli uczenia maszynowego.

## Zastosowania baz wektorowych
Bazy wektorowe znajdują zastosowanie w wielu dziedzinach:
* **Przetwarzanie języka naturalnego**: Umożliwiają analizę tekstu i generowanie odpowiedzi na podstawie kontekstu.
* **Systemy rekomendacyjne**: Pomagają w dostosowywaniu treści do preferencji użytkowników.
* **Analiza obrazów**: Umożliwiają klasyfikację i wyszukiwanie obrazów na podstawie ich cech wizualnych.

W miarę jak ilość i złożoność danych rośnie, bazy wektorowe stają się kluczowym narzędziem dla organizacji chcących efektywnie przetwarzać i analizować informacje

## Najpopularniejsze algorytmy wyszukiwania wektorowego
Wyszukiwanie wektorowe wykorzystuje różne algorytmy do efektywnego znajdowania podobnych wektorów w przestrzeni wielowymiarowej. Oto niektóre z najpopularniejszych algorytmów:
* **K-Najbliżsi Sąsiedzi (KNN)**: To jeden z najprostszych i najczęściej stosowanych algorytmów, który wykonuje wyszukiwanie siłowe, przeszukując całą przestrzeń wektorów w celu znalezienia K najbliższych sąsiadów do danego wektora zapytania.
  
* **Hierarchical Navigable Small World (HNSW)**: Algorytm ten realizuje przybliżone wyszukiwanie najbliższych sąsiadów (ANN) poprzez budowę grafu, który umożliwia szybkie przeszukiwanie i nawigację w przestrzeni wektorowej. HNSW jest znany z wysokiej wydajności i dokładności.
* **Faiss**: Opracowany przez Facebooka, Faiss to biblioteka do wydajnego wyszukiwania podobieństwa i grupowania gęstych wektorów. Umożliwia przetwarzanie dużych zbiorów danych, nawet tych, które nie mieszczą się w pamięci RAM.
* **NGT (Neighborhood Graph and Tree)**: Jest to algorytm, który wykorzystuje grafy do efektywnego wyszukiwania najbliższych sąsiadów. NGT jest znany z szybkości działania i zdolności do przetwarzania dużych zbiorów danych.
* **Annoy (Approximate Nearest Neighbors Oh Yeah)**: Stworzony przez Spotify, Annoy jest algorytmem, który buduje drzewo binarne dla efektywnego wyszukiwania. Jest szczególnie przydatny w aplikacjach wymagających szybkiego dostępu do danych.
* **LSH (Locality-Sensitive Hashing)**: Algorytm ten grupuje podobne wektory w "koszykach", co pozwala na szybkie wyszukiwanie podobieństw bez konieczności przeszukiwania całej przestrzeni wektorowej.
  

Każdy z tych algorytmów ma swoje unikalne cechy i zastosowania, co sprawia, że są one odpowiednie dla różnych scenariuszy wyszukiwania wektorowego. Wybór odpowiedniego algorytmu zależy od specyficznych wymagań aplikacji oraz charakterystyki danych.

# Przykład

## Wizualna reprezentacja modelu 2D

````python
from langchain_ollama import OllamaEmbeddings
from langchain_core.vectorstores import InMemoryVectorStore
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

documents = [
    "LangChain is a framework for building applications with LLMs.",
    "Ollama provides advanced LLM-based embeddings.",
    "Vector databases are essential for efficient text similarity search.",
    "FAISS is a popular library for vector similarity search.",
    "There is no man playing the piano with his nose."
]

embedding_model = OllamaEmbeddings(model="nomic-embed-text:latest")
vector_store = InMemoryVectorStore.from_texts(documents, embedding=embedding_model)

embeddings = vector_store.embedding
embeddings = np.array(embeddings.embed_documents(documents))

tsne = TSNE(n_components=2, random_state=42, perplexity=len(embeddings)-1)
embeddings_2d = tsne.fit_transform(embeddings)

plt.figure(figsize=(10, 7))
for i, label in enumerate(documents):
    plt.scatter(embeddings_2d[i, 0], embeddings_2d[i, 1], label=f"Doc {i+1}")
    plt.annotate(label[:30] + "...", (embeddings_2d[i, 0], embeddings_2d[i, 1]))


plt.title("2D Visualization of Embeddings")
plt.xlabel("Dimension 1")
plt.ylabel("Dimension 2")
plt.legend()
plt.show()
````

## Wizualna reprezentacja modelu 3D

````python
from langchain_ollama import OllamaEmbeddings
from langchain_core.vectorstores import InMemoryVectorStore
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

documents = [
    "LangChain is a framework for building applications with LLMs.",
    "Ollama provides advanced LLM-based embeddings.",
    "Vector databases are essential for efficient text similarity search.",
    "FAISS is a popular library for vector similarity search.",
    "There is no man playing the piano with his nose."
]

embedding_model = OllamaEmbeddings(model="nomic-embed-text:latest")
vector_store = InMemoryVectorStore.from_texts(documents, embedding=embedding_model)

embeddings = vector_store.embedding
embeddings = np.array(embeddings.embed_documents(documents))

tsne = TSNE(n_components=3, random_state=42, perplexity=len(embeddings)-1)
embeddings_3d = tsne.fit_transform(embeddings)

# Plot embeddings in 3D
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
for i, label in enumerate(documents):
    ax.scatter(embeddings_3d[i, 0], embeddings_3d[i, 1], embeddings_3d[i, 2], label=f"Doc {i+1}")
    ax.text(embeddings_3d[i, 0], embeddings_3d[i, 1], embeddings_3d[i, 2], label[:30] + "...", fontsize=8)

ax.set_title("3D Visualization of Embeddings")
ax.set_xlabel("Dimension 1")
ax.set_ylabel("Dimension 2")
ax.set_zlabel("Dimension 3")
plt.legend()
plt.show()
````

## Porównywanie tekstów

### Przeszukiwanie
Bazy danych osadzają i przechowują dodane dokumenty. Jeśli przekażemy zapytanie, Baza wektorów przy uzyciu modelu osadzi zapytanie, przeprowadzi wyszukiwanie podobieństw w osadzonych dokumentach i zwróci najbardziej podobne.

**Odzwierciedla to dwie ważne koncepcje:**
* Po pierwsze, musi istnieć sposób zmierzenia podobieństwa między zapytaniem a dowolnym osadzonym dokumentem.
* Po drugie, musi istnieć algorytm, który skutecznie przeprowadzi wyszukiwanie podobieństwa we wszystkich osadzonych dokumentach.

### Metryki podobieństwa
Krytyczną zaletą `embeddings vector` jest to, że można je porównywać za pomocą wielu prostych operacji matematycznych:

* **Podobieństwo cosinusowe:** Mierzy cosinus kąta między dwoma wektorami.
* **Odległość euklidesowa:** Mierzy odległość w linii prostej pomiędzy dwoma punktami.
* **Iloczyn skalarny:** Mierzy rzut jednego wektora na drugi.

#### Na później
> * Sprawdziź [dokumentacje](https://developers.google.com/machine-learning/clustering/dnn-clustering/supervised-similarity) z Google na temat metryk porównywania.
> * Sprawdzić Pinecone's [blog post](https://www.pinecone.io/learn/vector-similarity/) na temat metryk porównywania.

````python
from langchain_core.vectorstores import InMemoryVectorStore
from langchain_core.documents import Document
from langchain_ollama import OllamaEmbeddings


embedding_model = OllamaEmbeddings(model="nomic-embed-text:latest")

# Initialize with an embedding model
vector_store = InMemoryVectorStore(embedding=embedding_model)

document1 = Document(
    page_content="I had chocolate chip pancakes and scrambled eggs for breakfast this morning.",
    metadata={"source": "tweet"},
)

document2 = Document(
    page_content="The weather forecast for tomorrow is cloudy and overcast, with a high of 62 degrees.",
    metadata={"source": "news"},
)

vector_store.add_documents(documents=[document1, document2])

doc = vector_store.similarity_search_with_score(
    query="What weather will be tomorrow?", k=2
)
print("Content:", doc[0][0].page_content)
print("Score similarity:", doc[0][1])
````
