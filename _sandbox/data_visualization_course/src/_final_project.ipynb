## Setup

### Load datasets

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from pathlib import Path
import pandas as pd
pd.plotting.register_matplotlib_converters()

DC_FILE = Path("./data/dc-wikia-data.csv")
MARVEL_FILE = Path("./data/marvel-wikia-data.csv")

dc_df = pd.read_csv(DC_FILE)
marvel_df = pd.read_csv(MARVEL_FILE)
```

## Task Alive vs Deceased Characters

Compares the number of Living Characters versus those who are not.

```python
sns.set_theme(style='whitegrid')

alive_counts = dc_df['ALIVE'].value_counts()

plt.figure(figsize=(8, 6))
alive_counts.plot(kind='pie')

plt.title('Alive vs Deceased')
plt.xlabel('Status')
plt.ylabel('Character count')
plt.xticks(rotation=0)

plt.show()
```

```python
alive_counts = marvel_df['ALIVE'].value_counts()

plt.figure(figsize=(8, 6))
alive_counts.plot(kind='pie')

plt.title('Alive vs Deceased')
plt.xlabel('Status')
plt.ylabel('Character count')
plt.xticks(rotation=0)

plt.show()
```

## Task Character Alignment Distribution

Compares the number of Good Characters, Bad Characters, and Neutral Characters

```python
align_count = dc_df['ALIGN'].value_counts()
print(align_count)

align_count.plot(kind='pie')
plt.title('Align Characters')
plt.xlabel('Status')
plt.ylabel('Character count')
plt.show()
```

```python
sns.set_theme(style="whitegrid")

# Prepare data for bar plot
alignment_counts = dc_df['ALIGN'].value_counts().reset_index()
alignment_counts.columns = ['ALIGN', 'count']

# Create a bar plot for the distribution of characters by alignment
plt.figure(figsize=(10, 6))
ax = sns.barplot(data=alignment_counts, x='ALIGN', y='count')

# Add labels and title
plt.xlabel('Alignment')
plt.ylabel('Number of Characters')
plt.title('Distribution of DC Characters by Alignment')
plt.xticks(rotation=45)
```
